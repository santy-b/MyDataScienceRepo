# Shape Classes

This Jupyter notebook demonstrates how to create and use classes in Python to draw and manipulate shapes. The notebook creates three classes: Circle, Rectangle, and Square. Each class has methods for drawing the shape, changing its dimensions, and changing its color.

First we import the class to draw the objects.

```python
import matplotlib.pyplot as plt
%matplotlib inline  
```

## Circle Class

```python
class Circle(object):
    # Constructor
    def __init__(self, radius = 4, color = "red"):
        self.radius = radius
        self.color = color
        
        
    # Method's
    def add_radius(self, r):
        self.radius = self.radius + r
        return self.radius
    
    def change_color(self, c):
        self.color = c
        return self.color
    
    def draw_circle(self):
        plt.gca().add_patch(plt.Circle((0, 0), radius=self.radius, fc=self.color))
        plt.axis('scaled')
        plt.show()  
```

```python
circle = Circle(6, "blue")
```

```python
circle.draw_circle()
```

```python
circle.change_color("red")
circle.add_radius(100)
circle.draw_circle()
```

```python
print("radius =", circle.radius, "color =", circle.color)
```

## Rectangle Class

```python
class Rectangle(object):
    # Constructor
    def __init__(self, height=5, width=5, color="blue"):
        self.height = height
        self.width = width
        self.color = color
        
    # Method's
    def chage_size(self, h, w):
        self.height = h
        self.width = w
        
    def change_color(self, color):
        self.color = color
    
    def draw_rectangle(self):
        plt.gca().add_patch(plt.Rectangle((0, 0), self.width, self.height ,fc=self.color))
        plt.axis('scaled')
        plt.show()
```

```python
rectangle = Rectangle(5, 10, "yellow")
rectangle.draw_rectangle()
```

## Triangle Class 

```python
class Triangle(object):
    
    # Constructor
    def __init__(self, height=5, width=5, color="blue"):
        self.color = color
        self.height = height
        self.width = width
        self.vertices = [(0, 0), (self.width, 0), (self.width/2, self.height)]
        
    # Method's
    def chage_size(self, h, w):
        self.height = h
        self.width = w
        self.vertices = [(0, 0), (self.width, 0), (self.width/2, self.height)]
        
    def change_color(self, color):
        self.color = color
    
    def draw_triangle(self):
        plt.gca().add_patch(plt.Polygon(self.vertices, fc=self.color))
        plt.axis('scaled')
        plt.show()
```

```python
triangle = Triangle(10, 12, "red")
triangle.draw_triangle()
```

```python
triangle.chage_size(10, 12)
triangle.change_color("blue")
triangle.draw_triangle()
```
