```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

# Extracting Stock Data With Python

```python
apple = yf.Ticker("AAPL")
cisco = yf.Ticker("CSCO")
```

yfinance documentation https://aroussi.com/post/python-yahoo-finance

```python
apple_info = apple.info
cisco_info = cisco.info
print("Apple Information: ", apple_info)
print("Cisco Information: ", cisco_info)
```

## We can access a companies information

```python
cisco_info["country"]
```

## Extracting Share Prices

```python
apple_share_price_data = apple.history(period="max")
cisco_share_price_data = cisco.history(period="max")
```

The reset_index function in pandas is used to reset the index of a DataFrame. When you call reset_index, the current index of the DataFrame is replaced with a default integer index (0, 1, 2, ...), and the previous index values become a new column in the DataFrame.

```python
apple_share_price_data.reset_index(inplace=True)
cisco_share_price_data.reset_index(inplace=True)
print("Apple Share Price Data: ", apple_share_price_data.head())
print("Cisco Share Price Data: ", cisco_share_price_data.head())
```

## Plot the 'Open' values for both datasets

### Apple

```python
print(apple_share_price_data.plot(x="Date", y="Open"))
```

## Plot the 'Dividends Returned' for both datasets

### Cisco

```python
print(cisco_share_price_data.plot(x="Date", y="Open"))
```
